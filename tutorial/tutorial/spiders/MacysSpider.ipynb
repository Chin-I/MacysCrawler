```python

import scrapy
import re 
import time
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings

class MacysSpider(scrapy.Spider):
    name = "macysspider"
    start_urls = [
        'https://www.macys.com/',
    ]

    def parse(self, response):
        for href in response.css('li.fob a::attr(href)'):
            yield response.follow(href, self.parse_headers)

    def parse_headers(self, response):
        for href in response.css('li.children a::attr(href)'):
            yield response.follow(href, self.parse_links)

    def parse_links(self, response):
        for href in response.css('a.productDescLink::attr(href)'):
            yield response.follow(href, self.parse_products)
        next_page = response.css('div#filterResultsBottom ul li ul li.nextPage a::attr(href)').extract_first()

        if next_page is not None:
            f = open("out-macys-nextpage.txt","a+")
            f.write( next_page+"\n" )   
            f.close()
            yield response.follow(next_page, callback=self.parse)


    def parse_products(self, response):
        for product_name in response.css('div.productTitle'):
            yield {
                'product_name': product_name.css('.productName::text').extract_first()
                # product_name.css('.productName::text').extract_first()
            }
            pn = ( product_name.css('.productName::text').extract_first() )
            f = open("out-macys-productname.txt","a+")
            f.write( pn+"\n" )   
            f.close()

##################################################
start_time = time.time()
            
process = CrawlerProcess(get_project_settings())

process.crawl(MacysSpider)
process.start() # the script will block here until the crawling is finished

end_time = time.time() - start_time
print('[DONE] Elasped Time: {}'.format(end_time))
```
